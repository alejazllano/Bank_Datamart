```python
# Each record describes characteristics of a client
client = pd.read_csv('./data_berka/client.asc', sep=';')
client.head()
```

```python
# Getting year, day and month birthday and gender of clients

client['birth_year'] = client['birth_number'].transform(lambda bn: int('19' + str(bn)[:2]))
client['birth_day'] = client['birth_number'].astype(str).str[-2:].astype(int)

def to_month_gender(birth_number):
    
    s = str(birth_number)
    birth_month = int(s[2:4])
    
    if birth_month > 50:
        gender = "F"
        birth_month = birth_month - 50
    else:
        gender = 'M'
        
    return pd.Series({'birth_month':birth_month, 'gender':gender})

client[['birth_month', 'gender']] = client['birth_number'].apply(to_month_gender)
```

```python
# Dropping birth number

client=client.drop(['birth_number'], axis=1)
```

```python
# Calculating client's age

client.loc[:,'age']=1999-client['birth_year']
client.head()
```

```python
trans_agg = pd.DataFrame()
```

```python
trans_agg['account_id'] = trans['account_id'].unique()
```

```python
trans_agg.nunique()
```

```python
trans_agg = trans_agg.merge(trans[['account_id','amount']].groupby('account_id',as_index = False).agg(max),on = 'account_id', how = 'left')
```

```python
trans_agg[['account_id','amount']].sort_values('account_id')
```

```python
trans_agg = trans_agg.merge(trans[['account_id','amount']].groupby('account_id',as_index = False).agg(min),on = 'account_id', how = 'left')
```

```python
#getting aggregate statistics based on each account

trans_agg = trans_agg.merge(trans[['account_id','amount']].groupby('account_id',as_index = False).agg([('trans_amt_total','sum'),('trans_amt_min','min'),('trans_amt_max','max'),('trans_number','count')]),on = 'account_id',how='left')

```

```python
trans_agg
```

```python
def old_trans(id):
    a = min(trans[trans['account_id'] == id]['date_trans'])  
    days = (datetime.datetime(1999,1,1) - a).days
    return days
```

```python
#get the oldest transaction by date for each account_id
a=trans[['account_id','date_trans']].groupby('account_id',as_index = False).agg(min)
#merge it with the trans_agg table and rename the column
trans_agg=trans_agg.merge(a,on = 'account_id', how = 'left')
#get the number of  days since this date
trans_agg['days_oldtrans']=trans_agg['account_id'].apply(old_trans)
print(trans_agg.head())
```

```python
def latest_trans(id):
    a = max(trans[trans['account_id'] == id]['date_trans'])  
    days = (datetime.datetime(1999,1,1) - a).days
    return days
```

```python
#get the latest transaction by date for each account_id
a=trans[['account_id','date_trans']].groupby('account_id',as_index = False).agg(max)
#merge it with the trans_agg table and rename the column
trans_agg=trans_agg.merge(a,on = 'account_id', how = 'left')
#get the number of  days since this date
trans_agg['days_latetrans']=trans_agg['account_id'].apply(latest_trans)
print(trans_agg.head())
```

```python
trans_agg
```

```python
#get the latest balance by date for each account_id
a=trans[['account_id','date_trans','balance']].groupby('account_id',as_index = False).agg([('date_trans','max')])
#merge it with the trans_agg table and rename the column
trans_agg=trans_agg.merge(a,on = 'account_id', how = 'left')
trans_agg
```

```python
#get the average balance for each account_id
a=trans[['account_id','date_trans','balance']].groupby('account_id',as_index = False).agg([('balance','mean')])
#merge it with the trans_agg table and rename the column
trans_agg=trans_agg.merge(a,on = 'account_id', how = 'left')
trans_agg
```

```python
trans_agg.columns=['account_id','amount_x','amount_y',
          'trans_amt_total','trans_amt_min','trans_amt_max',
          'trans_number','date_trans_old','days_oldtrans',
          'date_trans_late','days_latetrans','date_balance',
           'latest_balance', 'avg_balance']
trans_agg
```

```python
trans_agg = trans_agg.drop(['amount_x','amount_y'], axis = 1)
```
